```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from concurrent.futures import ThreadPoolExecutor

# Download necessary resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Load dataset
file_path = "/content/Merged2.csv"
data = pd.read_csv(file_path)

# Ensure required columns exist
if 'Questions' not in data.columns or 'Response' not in data.columns:
    raise ValueError("Dataset must contain 'Questions' and 'Response' columns.")

# Initialize NLP tools
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Function to tokenize and clean text
def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', str(text))  # Remove punctuation
    words = word_tokenize(text.lower())  # Tokenize and convert to lowercase
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words]  # Remove stopwords & lemmatize
    return " ".join(words)  # Return cleaned text

# Apply tokenization using multi-threading
with ThreadPoolExecutor() as executor:
    data['Tokenized_Questions'] = list(executor.map(preprocess_text, data['Questions']))
    data['Tokenized_Response'] = list(executor.map(preprocess_text, data['Response']))

# Function to summarize responses using TF-IDF
def summarize_response(text, num_sentences=2):
    if pd.isna(text) or len(text.split()) < 10:
        return text  # Skip if too short

    sentences = nltk.sent_tokenize(text)
    if len(sentences) <= num_sentences:
        return text  # If already short, return as-is

    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(sentences)
    scores = tfidf_matrix.sum(axis=1).A1
    top_sentences = [sentences[i] for i in scores.argsort()[-num_sentences:]]
    return " ".join(top_sentences)

# Apply summarization
data['Summarized_Response'] = data['Response'].apply(summarize_response)

# Save processed data
data.to_csv("/content/Merged2_processed_fast.csv", index=False)
print("Processed dataset saved successfully!")
```

```python
import google.generativeai as genai
import os
import numpy as np
import nltk
import string
import random
import pandas as pd
import gradio as gr
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.stem import WordNetLemmatizer

# Set API key for Google Gemini API
os.environ["GOOGLE_API_KEY"] = "AIzaSyB72DIiTbIJrZKfMARNWV3N24HzDBR-U38"
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

# Load dataset
file_path = "Merged2_processed_fast.csv"
df = pd.read_csv(file_path)

# Ensure required columns exist
if "Questions" not in df.columns or "Summarized_Response" not in df.columns:
    raise KeyError("Dataset must have 'Questions' and 'Summarized_Response' columns.")

# Preprocess dataset
sent_tokens = df['Questions'].dropna().astype(str).str.lower().tolist()
lemmer = WordNetLemmatizer()
remove_punct_dict = str.maketrans("", "", string.punctuation)

def LemTokens(tokens):
    return [lemmer.lemmatize(token) for token in tokens]

def LemNormalize(text):
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))

# Setup TF-IDF vectorization (Train Once)
TfidfVec = TfidfVectorizer(tokenizer=LemNormalize, stop_words='english')
tfidf_matrix = TfidfVec.fit_transform(sent_tokens)

# Crisis keywords
CRISIS_KEYWORDS = {
    "suicide", "die", "end life", "kill myself", "want to die",
    "no reason to live", "hopeless", "i feel like disappearing"
}

HELPLINE_RESPONSE = (
    "I'm really sorry you're feeling this way. You're not alone, and support is available. "
    "Please consider reaching out to a mental health professional or a trusted person. If you need urgent help, "
    "please call a crisis helpline:\n\n"
    "**📞 National Suicide Prevention Helpline:** 988 (USA)\n"
    "**📞 Samaritans (UK):** 116 123\n"
    "**📞 Snehi (India):** +91-9582208181\n"
    "You are valuable, and there are people who care about you. Please reach out for help. 💙"
)

# Fetch response from Gemini API
def fetch_gemini_response(prompt):
    structured_prompt = f"""
    You are a compassionate mental health chatbot trained in psychology and medical science.
    Provide short, helpful, and medically accurate advice in **2-3 sentences**.

    - Acknowledge the user's feelings warmly.
    - Suggest a **specific**, actionable self-help strategy.
    - Offer emotional support and remind them they are not alone.
    - If the distress is severe, gently suggest seeking professional help.

    **User:** {prompt}
    **Chatbot:**
    """

    try:
        response = genai.GenerativeModel("gemini-2.0-flash").generate_content(structured_prompt)

        if not response or not hasattr(response, "text") or not response.text.strip():
            return "I'm sorry, but I couldn't generate a meaningful response. Please try rephrasing your question."

        # Ensure response is limited to 4-5 key sentences
        sentences = response.text.strip().split(". ")
        short_response = ". ".join(sentences[:5]) + "." if len(sentences) > 1 else response.text.strip()

        return short_response

    except ValueError:
        return "I'm sorry, but I couldn't process that request. Please try again."

# Hybrid chatbot function (Dataset + Gemini API)
def chatbot_response(message, history=None):  # Added history argument to match Gradio's signature
    message = message.lower()

    # Detect crisis and provide helpline immediately
    if any(keyword in message for keyword in CRISIS_KEYWORDS):
        return HELPLINE_RESPONSE

    # Handle basic greetings
    GREET_INPUTS = ("hello", "hi", "greetings", "sup", "what's up", "hey")
    GREET_RESPONSES = ["Hi!", "Hey!", "Hello!", "Howdy!", "Hi there!", "Hey, what's up?"]
    if any(word in message.split() for word in GREET_INPUTS):
        return random.choice(GREET_RESPONSES)

    # Handle farewells
    if message in ["bye", "exit", "quit"]:
        return "Goodbye! Take care!"
    elif message in ["thanks", "thank you"]:
        return "You're welcome! I'm always here to help."

    # NLP-based response using dataset
    user_tfidf = TfidfVec.transform([message])
    vals = cosine_similarity(user_tfidf, tfidf_matrix)
    idx = vals.argsort()[0][-1]  # Get the best matching index
    flat = vals.flatten()
    flat.sort()
    req_tfidf = flat[-1]

    # If dataset confidence is high, return dataset response
    SIMILARITY_THRESHOLD = 0.6
    if req_tfidf > SIMILARITY_THRESHOLD:
        response = df['Summarized_Response'].iloc[idx]

        # Prevent irrelevant responses (like "Crying is a healthy way...")
        if "crying" in response.lower() or "release emotions" in response.lower():
            return fetch_gemini_response(message)  # Fetch better response from API

        return "\n".join(response.split("\n")[:4])  # Limit response to 4 lines

    # Otherwise, use Gemini API
    return fetch_gemini_response(message)

# Gradio UI setup
with gr.Blocks() as app:
    gr.Markdown("# 🧠Thrive Mental Health Chatbot 🤖")
    gr.Markdown("A chatbot for mental well-being support. ")

    chatbot = gr.ChatInterface(fn=chatbot_response, title="Thrive Mental Health Chatbot", type="messages")

    gr.Markdown("⚠️ This chatbot provides general advice. Seek professional help for serious issues.")

app.launch(share=True, debug=True)
```
